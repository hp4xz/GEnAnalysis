```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.colors import LogNorm
from matplotlib.colors import PowerNorm

# Define the custom colormap and reverse the order
colors = ['cyan', '#f2be44','red']
colors = ['#000004', '#b63679', '#fc9662', '#fcffa4']
colors = ['white', 'red','#fc9662','orange','yellow']
cmap = mcolors.LinearSegmentedColormap.from_list('custom_cmap', colors)

# Generate random data
x = np.random.randn(1000000)
y = np.random.randn(1000000)

# Create 2D histogram
plt.figure(figsize=(8, 6))
hist = plt.hist2d(x, y, bins=200, cmap=cmap,norm=PowerNorm(gamma=.3))

# Add color bar
plt.colorbar(hist[3], label='Counts')

# Customize the plot
plt.title('2D Histogram with Inverted Custom Color Scheme')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()

```
