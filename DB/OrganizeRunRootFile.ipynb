```python
import ROOT
import csv
# Open the ROOT file
file = ROOT.TFile("_GEN2RunDates.root")

# Access the tree
tree = file.Get("Tout")

# Dictionary to store the first and last datetime for each runnum
run_dates = {}

# Function to format datetime as a string
def format_datetime(dt):
    year = dt.GetYear()
    month = dt.GetMonth()
    day = dt.GetDay()
    hour = dt.GetHour()
    minute = dt.GetMinute()
    return f"{year}-{month:02d}-{day:02d} {hour:02d}:{minute:02d}"

# Loop through the tree
for entry in tree:
    runnum = entry.runnum
    datetime = format_datetime(entry.datetime)

    if runnum not in run_dates:
        run_dates[runnum] = {"first": datetime, "last": datetime}
    else:
        run_dates[runnum]["last"] = datetime

# Print the results
for runnum, dates in run_dates.items():
    print(f"Run Number: {runnum}, First datetime: {dates['first']}, Last datetime: {dates['last']}")

# Close the file
file.Close()
```

```python
import ROOT
import pandas as pd

# List of ROOT files
root_files = ["_GEN2RunDates.root", "_GEN3RunDates.root", "_GEN4RunDates.root","_GEN4bRunDates.root"]

# Dictionary to store the first and last datetime for each runnum
run_dates = {}

# Function to format datetime as a string
def format_datetime(dt):
    year = dt.GetYear()
    month = dt.GetMonth()
    day = dt.GetDay()
    hour = dt.GetHour()
    minute = dt.GetMinute()
    return f"{year}-{month:02d}-{day:02d} {hour:02d}:{minute:02d}"

# Process each ROOT file
for root_file in root_files:
    file = ROOT.TFile(root_file)
    tree = file.Get("Tout")
    
    for entry in tree:
        runnum = entry.runnum
        datetime = format_datetime(entry.datetime)
        
        if runnum not in run_dates:
            run_dates[runnum] = {"first": datetime, "last": datetime}
        else:
            run_dates[runnum]["last"] = datetime
    
    file.Close()

# Convert the run_dates dictionary to a pandas DataFrame
df = pd.DataFrame([
    {"Run": runnum, "Start": dates["first"], "Finish": dates["last"]}
    for runnum, dates in run_dates.items()
])

# Sort the DataFrame by Run number
df.sort_values(by="Run", inplace=True)

# Write the DataFrame to a CSV file
df.to_csv("RunDatesV3.csv", index=False)
```

```python
with open("RunDatesV2.csv", "w", newline='') as csvfile:
    fieldnames = ["Run", "Start", "Finish"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for runnum, dates in run_dates.items():
        writer.writerow({"Run": runnum, "Start": dates['first'], "Finish": dates['last']})

```
